### 3.7 Dodatkowe informacje o danych

```python
# 1
sql = """ SELECT 
    SUM(new_confirmed) AS total_confirmed_cases,
    SUM(new_deceased) AS total_deaths,
    (SUM(new_deceased)/SUM(new_confirmed))*100 AS case_fatality_ratio
    FROM `bigquery-public-data.covid19_open_data.covid19_open_data`
    WHERE country_name='Poland'
         """
q1 = (sql)
query_job = client.query(q1)
query_result = query_job.result()

query_result.to_dataframe()
```

```python
# 2 populacja i gęstość zaludnienia

sql = """

        select
            date,
            population_density,
            population
        from `bigquery-public-data.covid19_open_data.covid19_open_data`
        where country_name='Poland'
            and aggregation_level = 0
        limit 10

"""
q1 = (sql)
query_job = client.query(q1)
query_result = query_job.result()

query_result.to_dataframe()
```

```python
# 3 zmiany mobility

sql = """
        select
            date,
            location_key,
            new_confirmed,
            mobility_retail_and_recreation,
            mobility_grocery_and_pharmacy,
            mobility_parks,
            mobility_transit_stations,
            mobility_workplaces,
            mobility_residential
        from `bigquery-public-data.covid19_open_data.covid19_open_data`
        where aggregation_level = 1
        limit 10
"""
q1 = (sql)
query_job = client.query(q1)
query_result = query_job.result()

query_result.to_dataframe()
```

```python
# 4 szerokość geograficzna i wysokość nad poziomeme morza porównana z liczbą potwierdzonych

sql = """
        select
            location_key,
            date,
            cumulative_confirmed,
            latitude,
            longitude,
            elevation_m
        from `bigquery-public-data.covid19_open_data.covid19_open_data`
        where rainfall_mm is not null
            and snowfall_mm is not null
            and aggregation_level = 0
        limit 10
"""
q1 = (sql)
query_job = client.query(q1)
query_result = query_job.result()

query_result.to_dataframe()
```

```python
# 5 ilość dni z zamniętymi szkołami / miejscami pracy / imprezami

sql = """
with t as (
        select
            location_key,
            school_closing
        from `bigquery-public-data.covid19_open_data.covid19_open_data`
        where school_closing >= 2
            and aggregation_level = 0
) select location_key,
         count(school_closing) closed_school_days
  from t
  group by location_key
  limit 10
"""
q1 = (sql)
query_job = client.query(q1)
query_result = query_job.result()

query_result.to_dataframe()
```

```python
# 6 tredy wyszukiwań google najweksze wyszukiwanie bólu klatki

sql = """
select
    date,
    location_key,
    search_trends_burning_chest_pain
from `bigquery-public-data.covid19_open_data.covid19_open_data`
where search_trends_burning_chest_pain is not null
    and aggregation_level = 0
order by 3
limit 15
"""

q1 = (sql)
query_job = client.query(q1)
query_result = query_job.result()

query_result.to_dataframe()
```

```python
# 4.1 podstawowe dane o wszystkich krajach świata
import numpy as np

q1 = ("""select distinct country_name `nazwakaraju`,
         location_key `kod`, 
         iso_3166_1_alpha_2, iso_3166_1_alpha_3,
         physicians_per_1000 `psycholodzy_na_1000_mieszkancow`,
         population `populacja`,
         population_male `mezczyzni`,
         population_female `kobiety`,
         gdp_usd,
         area_sq_km
         from `bigquery-public-data.covid19_open_data.covid19_open_data`
         where country_name is not null
             and aggregation_level = 0
             
         """)
query_job = client.query(q1)
query_result = query_job.result()
df = query_result.to_dataframe().dropna().drop_duplicates()

df.to_csv("kraje.csv", index=False)
df
```

```python
# 4.2 statyski światowe

q1 = ("select country_code, sum(new_confirmed) total_confirmed, sum(new_deceased) total_deaths, sum(new_recovered) total_recovers, sum(new_intensive_care_patients) total_intensive_patients from `bigquery-public-data.covid19_open_data.covid19_open_data` group by country_code ")
query_job = client.query(q1)
query_result = query_job.result()
df = query_result.to_dataframe()
df.drop_duplicates()
out = df[['total_confirmed', "total_deaths", "total_recovers", "total_intensive_patients" ]].sum()
out.to_csv("4.2.csv")
out
```

```python
# 4.3 efekty śmiertelności
q1 = ("select country_code, extract(year from date) `year`, avg(cumulative_deceased) `max_deaths`, avg(human_development_index) `avg_hdi`, avg(gdp_per_capita_usd) `gdppc`, avg(nurses_per_1000) `nurses_per_1000`, avg(physicians_per_1000) `physicians_per_1000`   from `bigquery-public-data.covid19_open_data.covid19_open_data` group by 1, 2 ")
query_job = client.query(q1)
query_result = query_job.result()
df = query_result.to_dataframe()
```
