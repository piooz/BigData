### 3.7 Dodatkowe informacje o danych

```python
# 1 podsumowanie śmiertelności w Polsce
sql = """ SELECT 
    SUM(new_confirmed) AS total_confirmed_cases,
    SUM(new_deceased) AS total_deaths,
    (SUM(new_deceased)/SUM(new_confirmed))*100 AS case_fatality_ratio
    FROM `bigquery-public-data.covid19_open_data.covid19_open_data`
    WHERE country_name='Poland'
         """
q1 = (sql)
query_job = client.query(q1)
query_result = query_job.result()

query_result.to_dataframe()
```

```python
#3.7.2 Dzienne wyszukiwania 'kaszel' dla USA 
sql = """ SELECT date, avg(search_trends_cough) as avg_daily_search_cough
    FROM `bigquery-public-data.covid19_open_data.covid19_open_data`
    WHERE country_name='United States of America' and search_trends_cough IS NOT NULL
    group by date
    order by date desc
         """
q1 = (sql)
query_job = client.query(q1)
query_result = query_job.result()
query_result.to_dataframe()
```

```python
# 3 najwiekszy spadek mobility_workplaces w USA

sql = """
select
    date,
    location_key,
    new_confirmed,
    CAST(mobility_workplaces AS INT64) mobility_workplaces
from `bigquery-public-data.covid19_open_data.covid19_open_data`
    where aggregation_level = 1
        and country_name = 'United States of America'
        and CAST(mobility_workplaces AS INT64) = (select min(CAST(mobility_workplaces AS INT64))
                                    from bigquery-public-data.covid19_open_data.covid19_open_data
                                    where aggregation_level = 1
                                        and country_name = 'United States of America')
"""

q1 = (sql)
query_job = client.query(q1)
query_result = query_job.result()

query_result.to_dataframe()
```

```python
#3.7.4 Kraj gdzie stosunek nowych potwierdzonych do populacji jest najwiekszy
sql = """ 
SELECT country_name, new_confirmed/population AS pop_confirmed_ratio
    FROM (SELECT country_name, SUM(new_confirmed) AS new_confirmed, population
    FROM `bigquery-public-data.covid19_open_data.covid19_open_data`
    group by country_name,population
)
where new_confirmed/population = (SELECT MAX(ratio) 
    FROM (SELECT country_name, SUM(new_confirmed)/population AS ratio
    FROM `bigquery-public-data.covid19_open_data.covid19_open_data`
    group by country_name,population
))


"""
q1 = (sql)
query_job = client.query(q1)
query_result = query_job.result()
query_result.to_dataframe()
```

```python
# 5 ilość dni z zamniętymi szkołami / miejscami pracy / imprezami

sql = """
with t as (
        select
            location_key,
            school_closing
        from `bigquery-public-data.covid19_open_data.covid19_open_data`
        where school_closing >= 2
            and aggregation_level = 0
) select location_key,
         count(school_closing) closed_school_days
  from t
  group by location_key
"""
q1 = (sql)
query_job = client.query(q1)
query_result = query_job.result()

query_result.to_dataframe()
```

```python
# 3.7.6 top 5 hdi krajow
sql = """
select distinct
    location_key,
    country_name,
    human_development_index
from `bigquery-public-data.covid19_open_data.covid19_open_data`
where search_trends_burning_chest_pain is not null
    and aggregation_level = 0
    and human_development_index is not null
order by 3
limit 5
"""

q1 = (sql)
query_job = client.query(q1)
query_result = query_job.result()

query_result.to_dataframe()
```

```python
# 4.1 podstawowe dane o wszystkich krajach świata
import numpy as np

q1 = ("""select distinct country_name `nazwakaraju`,
         location_key `kod`, 
         iso_3166_1_alpha_2, iso_3166_1_alpha_3,
         physicians_per_1000 `psycholodzy_na_1000_mieszkancow`,
         population `populacja`,
         population_male `mezczyzni`,
         population_female `kobiety`,
         population_density gestosc_zaludnienia,
         gdp_usd,
         area_sq_km
         from `bigquery-public-data.covid19_open_data.covid19_open_data`
         where country_name is not null
             and aggregation_level = 0
             
         """)
query_job = client.query(q1)
query_result = query_job.result()
df = query_result.to_dataframe().dropna().drop_duplicates()

df.to_csv("kraje.csv", index=False)
df
```

```python
# 4.2 statyski światowe

q1 = ("select country_code, sum(new_confirmed) total_confirmed, sum(new_deceased) total_deaths, sum(new_recovered) total_recovers, sum(new_intensive_care_patients) total_intensive_patients from `bigquery-public-data.covid19_open_data.covid19_open_data` group by country_code ")
query_job = client.query(q1)
query_result = query_job.result()
df = query_result.to_dataframe()
df.drop_duplicates()
out = df[['total_confirmed', "total_deaths", "total_recovers", "total_intensive_patients" ]].sum()
out.to_csv("4.2.csv")
out
```

```python
# 4.3 efekty śmiertelności
q1 = ("""
select country_code, 
extract(year from date) `year`,
extract(month from date) `month`,
sum(new_deceased) `deaths`, 
sum(new_confirmed) confirmed,
(sum(new_deceased) / sum(new_confirmed)) * 100 mortality_rate
from `bigquery-public-data.covid19_open_data.covid19_open_data`
where new_confirmed > 1
group by 1, 2, 3
""")
query_job = client.query(q1)
query_result = query_job.result()
df = query_result.to_dataframe()

df = df.dropna().drop_duplicates()
df = df.sort_values(by=['country_code', 'year'])
df.to_csv("4.3.csv")
df
```
