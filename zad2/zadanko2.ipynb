# Część 3

```python
# inflacja (interesuje nas tylko 3 lata przed wybuchem pandemi)
inflacja = pd.read_csv('inflation.csv').drop(columns=['Indicator Type', 'Series Name', 'IMF Country Code', 'Unnamed: 59', 'Unnamed: 60', 'Unnamed: 61', 'Unnamed: 62', 'Unnamed: 63', 'Note'])
inflacja = inflacja.drop(inflacja.iloc[:, 2:49], axis=1)
#inflacja
inflacja = inflacja[['Country Code', '2017', '2018', '2019']]
            .add_prefix("inflation ")
            .groupby('inflation Country Code').max()
inflacja
```

```python
alpha = pd.merge(haos, inflacja,
         left_on=['CCA3'],
         right_on=['inflation Country Code'],
         how='inner')
alpha
```

```python
# zanieczyszczenia 

pollution = pd.read_csv("pollution.csv")
pollution
vals = ['Global', 'International Transport']
pollution = pollution[pollution.Country.isin(vals) == False]
pollution
```

```python
beta = pd.merge(alpha, pollution,
         left_on=['CCA3'],
         right_on=['ISO 3166-1 alpha-3'],
         how='inner')

beta
```

```python
# wartośc eksportu
export = pd.read_csv('export_info.csv')
# export
export = export.loc[export['Year'] == 2020]
export = export.groupby(by=['ReporterISO3', 'ReporterName', 'Year']).sum(numeric_only=True).drop_duplicates()
export
```

```python
gamma = pd.merge(beta, export,
         left_on=['CCA3', 'Year'],
         right_on=['ReporterISO3', 'Year'],
         how='inner')

gamma = gamma.drop(columns=['Unnamed: 0', 'ISO 3166-1 alpha-3', 'Country'])
gamma.to_csv('zad2.3.csv', index=False)
gamma
```
