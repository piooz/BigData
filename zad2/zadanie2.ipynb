```python
wc = pd.read_csv('world_countries.csv')
wc
```

```python
# inne selecja realnie potrzebnych informacji demograficznych (nieużywane)

inna = wc.melt(id_vars=['CCA3', 'Rank', 'Country/Territory', 'Capital', 'Continent', 'Area (km²)', "Density (per km²)", 'Growth Rate', 'World Population Percentage'],
        var_name='Year',
        value_name="Population" )

inna["Year"] = inna["Year"].str.split().str[0]
inna["Year"] = inna["Year"].astype(int)

#inna.rename({"Year": "Pop_Year"})

inna
```

```python
## łacznenie 2 nowych plików
#newdata = pd.merge(inna, gdp, left_on=['Country/Territory', 'CCA3', 'Year'], right_on=['Country Name', 'Country Code', 'Year'], how='left')
#newdata

```

```python
################# łączonko
fl["Year"] = fl["date"].str.split('-').str[0].astype(int)

haos = pd.merge(fl, gdp, how="left", left_on=['country_name', 'iso_3166_1_alpha_3', 'Year'], right_on=['Country Name', 'Country Code', 'Year'])
haos = haos.drop(columns=['Country Name', 'Country Code'])

haos = pd.merge(haos, inna, how='left', left_on=['country_name', 'Year'], right_on=['Country/Territory', 'Year'])
haos = haos.drop(columns=['Country/Territory', 'CCA3'])
haos['Year'] = haos['Year'].astype(str)
haos.to_csv('zad2.2.csv', index=False)
haos
```

# Część 3

```python
inflacja = pd.read_csv('inflation.csv')
inflacja
```

```python
alpha = pd.merge(haos, inflacja,
         left_on=['iso_3166_1_alpha_3', 'Year'],
         right_on=['Country Code', 'Year'],
         how='left')
alpha
```

```python
# zanieczyszczenia 

pollution = pd.read_csv("pollution.csv")

vals = ['Global', 'International Transport']
pollution = pollution[pollution.Country.isin(vals) == False]
pollution = pollution.drop(columns=['Unnamed: 0', 'Other', 'Country'])
pollution['Year'] = pollution['Year'].astype(str)
pollution
```

```python
beta = pd.merge(alpha, pollution,
         left_on=['iso_3166_1_alpha_3', 'Year'],
         right_on=['ISO 3166-1 alpha-3', 'Year'],
         how='left')

beta
```

```python
# wartośc eksportu
export = pd.read_csv('export_info.csv')
export['Year'] = export['Year'].astype(str)
# export
#export = export.loc[export['Year'] == 2020]
export = export.groupby(by=['ReporterISO3', 'ReporterName', 'Year']).sum(numeric_only=True).drop_duplicates()
# export.drop(columns='ReporterName')
export = export.reset_index()
export
```

```python
gamma = pd.merge(beta, export,
         left_on=['iso_3166_1_alpha_3', 'Year'],
         right_on=['ReporterISO3', 'Year'],
         how='left')

gamma = gamma.drop(columns=['ISO 3166-1 alpha-3'])
gamma.to_csv('zad2.3.csv', index=False)
gamma
```
