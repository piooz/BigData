```python
wc = pd.read_csv('world_countries.csv')
wc
```

```python
# selecja realnie potrzebnych informacji demograficznych

inna = wc.melt(id_vars=['CCA3', 'Rank', 'Country/Territory', 'Capital', 'Continent', 'Area (km²)', "Density (per km²)", 'Growth Rate', 'World Population Percentage'],
        var_name='Year',
        value_name="Population" )
inna["Year"] = inna["Year"].str.split().str[0]
inna["Year"] = inna["Year"].astype(int)
inna

# dane tylko dla roku 2020
inna = inna.loc[inna['Year'] == 2020].drop(columns='Year')
inna
```
