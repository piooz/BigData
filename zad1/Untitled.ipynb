```python
# 3.5 liczba dni bez czepień
q1 = ("select count(distinct date) from `bigquery-public-data.covid19_open_data.covid19_open_data` a where 0 = (select sum(new_vaccine_doses_administered) from `bigquery-public-data.covid19_open_data.covid19_open_data` where date = a.date)")
query_job = client.query(q1)
query_result = query_job.result()
query_result.to_dataframe()
```

```python
# 3.5 liczba dni bez zmarlych

sql = """select count(distinct date) 
        from `bigquery-public-data.covid19_open_data.covid19_open_data` a 
        where 0 = (
            select sum(new_deceased) 
            from `bigquery-public-data.covid19_open_data.covid19_open_data` 
            where date = a.date
        )"""
q1 = (sql)
query_job = client.query(q1)
query_result = query_job.result()
query_result.to_dataframe()
```

```python
# 3.6 zapisywanie dziennych informacji dla kraju? TODO:

sql = """select location_key, date, max(cumulative_confirmed) `cumulative_confirmed`
         from `bigquery-public-data.covid19_open_data.covid19_open_data` a
         where location_key = 'RO'
         group by date, location_key
         order by date
         """
q1 = (sql)
query_job = client.query(q1)
query_result = query_job.result()

query_result.to_dataframe()
```

```python
# 3.7 Dodatkowe informacje o danych

sql = """ SELECT SUM(cumulative_confirmed) AS total_confirmed_cases, SUM(cumulative_deceased) AS total_deaths, (SUM(cumulative_deceased)/SUM(cumulative_confirmed))*100 AS case_fatality_ratio
    FROM `bigquery-public-data.covid19_open_data.covid19_open_data`
    WHERE country_name='Poland'
         """
q1 = (sql)
query_job = client.query(q1)
query_result = query_job.result()

query_result.to_dataframe()
```

## Część 4

```python
# 4.1
import numpy as np

q1 = ("select distinct country_name `nazwakaraju`, country_code `kod`, iso_3166_1_alpha_2, iso_3166_1_alpha_3 from `bigquery-public-data.covid19_open_data.covid19_open_data` where country_name is not null")
query_job = client.query(q1)
query_result = query_job.result()
df = query_result.to_dataframe()
df.dropna().drop_duplicates().to_csv("kraje.csv", index=False)
```

```python
# 4.2 statyski światowe

q1 = ("select country_code, sum(new_confirmed) total_confirmed, sum(new_deceased) total_deaths, sum(new_recovered) total_recovers, sum(new_intensive_care_patients) total_intensive_patients from `bigquery-public-data.covid19_open_data.covid19_open_data` group by country_code ")
query_job = client.query(q1)
query_result = query_job.result()
df = query_result.to_dataframe()
df.drop_duplicates()
out = df[['total_confirmed', "total_deaths", "total_recovers", "total_intensive_patients" ]].sum()
out.to_csv("4.2.csv")
out
```

```python
# 4.3 efekty śmiertelności
q1 = ("select country_code, extract(year from date) `year`, avg(cumulative_deceased) `max_deaths`, avg(human_development_index) `avg_hdi`, avg(gdp_per_capita_usd) `gdppc`, avg(nurses_per_1000) `nurses_per_1000`, avg(physicians_per_1000) `physicians_per_1000`   from `bigquery-public-data.covid19_open_data.covid19_open_data` group by 1, 2 ")
query_job = client.query(q1)
query_result = query_job.result()
df = query_result.to_dataframe()
```
